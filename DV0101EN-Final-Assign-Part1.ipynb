````python
# Convertir la columna 'Date' a tipo datetime si aún no lo has hecho
df['Date'] = pd.to_datetime(df['Date'])

# Agrupar por año y calcular el promedio de ventas de automóviles
yearly_sales = df.groupby('Year')['Automobile_Sales'].mean().reset_index()

# Crear un gráfico de líneas utilizando pandas
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(yearly_sales['Year'], yearly_sales['Automobile_Sales'], marker='o', color='b', label='Ventas de Automóviles')

# Personalización del gráfico
ax.set_title('Automobile Sales during Recession', fontsize=16)
ax.set_xlabel('Año', fontsize=12)
ax.set_ylabel('Promedio de Ventas de Automóviles', fontsize=12)
ax.grid(True)

# Añadir marcas en el eje X con todos los años
ax.set_xticks(yearly_sales['Year'])

# Años de recesión
recession_years = [1980, 1981, 1991, 2000, 2007, 2020]

# Añadir anotaciones para los años de recesión
for year in recession_years:
    if year in yearly_sales['Year'].values:
        ax.annotate(f'Recesión {year}', 
                    xy=(year, yearly_sales.loc[yearly_sales['Year'] == year, 'Automobile_Sales'].values[0]), 
                    xytext=(year, yearly_sales['Automobile_Sales'].max() * 0.95),
                    arrowprops=dict(facecolor='red', shrink=0.05),
                    fontsize=10, color='red', ha='center')

# Guardar el gráfico como imagen
plt.savefig('line_plot_1_with_annotations.png')

# Mostrar el gráfico
plt.show()
````

####  TASK 1.2: Plot different lines for categories of vehicle type and analyse the trend to answer the question Is there a noticeable difference in sales trends between different vehicle types during recession periods?

````python
# Filtrar el DataFrame para obtener solo los registros durante el período de recesión
df_recession = df[df['Recession'] == 1]

# Agrupar por Año y Tipo de Vehículo, calculando el promedio de ventas
vehicle_sales_recession = df_recession.groupby(['Year', 'Vehicle_Type'], as_index=False)['Automobile_Sales'].mean()

# Crear un gráfico con líneas separadas para cada tipo de vehículo
plt.figure(figsize=(12, 8))

# Usar un ciclo para trazar una línea para cada tipo de vehículo
for vehicle in vehicle_sales_recession['Vehicle_Type'].unique():
    vehicle_data = vehicle_sales_recession[vehicle_sales_recession['Vehicle_Type'] == vehicle]
    plt.plot(vehicle_data['Year'], vehicle_data['Automobile_Sales'], label=vehicle, marker='o')

# Personalizar el gráfico
plt.title('Automobile Sales during Recession by Vehicle Type', fontsize=16)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Promedio de Ventas de Automóviles', fontsize=12)
plt.legend(title='Tipo de Vehículo')
plt.grid(True)

# Guardar el gráfico como imagen
plt.savefig('line_plot_vehicle_type_recession.png')

# Mostrar el gráfico
plt.show()
````

### TASK 1.3: Use the functionality of **Seaborn Library** to create a visualization to compare the sales trend per vehicle type for a recession period with a non-recession period.


<details><summary>Click here for a hint</summary>

<p>
     To visualize the average number of vehicles sold during recession and non-recession periods, you can use a bar chart
        <br> You will need to group recession average Automobile_Sales and then plot it<br>
    Make use of sns.barplot(x=x,y=y, data = df)</p>
</details>


````python
# Filter the data for recession and non-recession periods
recession_data = df[df['Recession'] == 1]
non_recession_data = df[df['Recession'] == 0]

# Group by vehicle type and calculate the average automobile sales for recession and non-recession periods
recession_sales = recession_data.groupby('Vehicle_Type')['Automobile_Sales'].mean().reset_index()
non_recession_sales = non_recession_data.groupby('Vehicle_Type')['Automobile_Sales'].mean().reset_index()

# Merge the two datasets for comparison
recession_sales['Recession'] = 'Recession'
non_recession_sales['Recession'] = 'Non-Recession'

# Combine both datasets
combined_sales = pd.concat([recession_sales, non_recession_sales])

# Create a bar plot using Seaborn
plt.figure(figsize=(10,6))
sns.barplot(x='Vehicle_Type', y='Automobile_Sales', hue='Recession', data=combined_sales)

# Adding labels and title
plt.title('Average Vehicle Sales During Recession and Non-Recession Periods', fontsize=16)
plt.xlabel('Vehicle Type', fontsize=12)
plt.ylabel('Average Sales', fontsize=12)
plt.xticks(rotation=45)
plt.show()
````

### TASK 1.4: Use sub plotting to compare the variations in GDP during recession and non-recession period by developing line plots for each period.
<br>Now, you want to find more insights from the data to understand the reason. <br>Plot a two line charts using subplotting to answer:-
#### How did the GDP vary over time during recession and non-recession periods? 
<br>Make use of <u>add_subplot()</u> from Matplotlib for this comparision.


````python
# Filter the data for recession and non-recession periods
recession_data = df[df['Recession'] == 1]
non_recession_data = df[df['Recession'] == 0]

# Create the figure with 2 subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Plot the GDP for recession period
ax1.plot(recession_data['Year'], recession_data['GDP'], color='red', label='Recession GDP')
ax1.set_title('GDP during Recession Period')
ax1.set_xlabel('Year')
ax1.set_ylabel('GDP')
ax1.grid(True)
ax1.legend()

# Plot the GDP for non-recession period
ax2.plot(non_recession_data['Year'], non_recession_data['GDP'], color='blue', label='Non-Recession GDP')
ax2.set_title('GDP during Non-Recession Period')
ax2.set_xlabel('Year')
ax2.set_ylabel('GDP')
ax2.grid(True)
ax2.legend()

# Display the plots
plt.tight_layout()
plt.show()
````

````python
# Filter data for non-recession years
non_recession_data = df[df['Recession'] == 0]

# Group the data by Month and calculate average sales
monthly_sales = non_recession_data.groupby('Month').agg(
    avg_sales=('Automobile_Sales', 'mean'),
    seasonality_weight=('Seasonality_Weight', 'mean')
).reset_index()

# Create a bubble plot
plt.figure(figsize=(10, 6))
plt.scatter(
    x=monthly_sales['Month'], 
    y=monthly_sales['avg_sales'], 
    s=monthly_sales['seasonality_weight'] * 100, # Size of bubble, multiplied for better visibility
    color='blue', 
    alpha=0.5
)

# Add title and labels
plt.title('Seasonality Impact on Automobile Sales')
plt.xlabel('Month')
plt.ylabel('Average Automobile Sales')

# Customize the x-axis to show month names
month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(range(1, 13), month_names)

# Show grid and plot
plt.grid(True)
plt.show()
````
